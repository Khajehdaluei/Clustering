```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from sklearn.cluster import SpectralClustering
```

```python
np.random.seed(0)
random_state = 170
n_samples = 1500

X_c, _ = datasets.make_circles(n_samples=n_samples, factor=0.5, noise=0.05)

X_m, _ = datasets.make_moons(n_samples=n_samples, noise=0.05)

X_s, _ = np.random.rand(n_samples, 2), None

X_b, _ = datasets.make_blobs(n_samples=n_samples, random_state=8)

X_v, _ = datasets.make_blobs(n_samples=n_samples, cluster_std=[1.0, 2.5, 0.5], random_state=random_state)

X, y = datasets.make_blobs(n_samples=n_samples, random_state=random_state)
transformation = [[0.6, -0.6], [-0.4, 0.8]]
X_aniso = np.dot(X, transformation)
X_a, _ = (X_aniso, y)

X_Data = [X_c, X_m, X_s, X_b, X_v, X_a]
```

```python
figure, axis = plt.subplots(1, 6,figsize=(25,5))
for j,i in enumerate(X_Data):
    X = StandardScaler().fit_transform(i)
    spectral = SpectralClustering(n_clusters=3,random_state=1,eigen_solver="arpack",affinity="nearest_neighbors")
    spectral.fit(X)
    y = spectral.labels_
    axis[j].scatter(i[:,0] ,i[:,1],c=y,)
```

```python
silhouette_score(X, spectral.labels_)
```
